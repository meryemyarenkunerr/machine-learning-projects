## *Train Test Split*

```python
X = df.drop("class", axis= 1)
y = df["class"]
```

```python
X = pd.get_dummies(X, drop_first= True)
```

```python
X.columns
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.1, random_state= 9)
```

## *AdaBoost Model*

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model = AdaBoostClassifier(n_estimators= 4)
```

```python
model.fit(X_train, y_train)
```

```python
mushroom_pred = model.predict(X_test)
```

## *Performance*

```python
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay, classification_report
```

```python
ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
```

```python
print(classification_report(y_test, mushroom_pred))
```

```python
error_rate = []

for mushroom_est in range(1,30):
    model = AdaBoostClassifier(n_estimators= mushroom_est)
    model.fit(X_train, y_train)
    mushroom_pred = model.predict(X_test)
    

    error = 1 - accuracy_score(y_test, mushroom_pred)
    error_rate.append(error)
```

```python
plt.plot(range(1,30), error_rate)
```

```python
model = AdaBoostClassifier(n_estimators= 16)
model.fit(X_train, y_train)
mushroom_pred = model.predict(X_test)

ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
print(classification_report(y_test, mushroom_pred))
```

```python
model.feature_importances_
```

```python
feature_df = pd.DataFrame(index= X.columns, columns= ["coefficients"], data= model.feature_importances_)
feature_df = feature_df[feature_df["coefficients"] > 0].sort_values("coefficients")
feature_df
```

```python
sns.barplot(data= feature_df, x= feature_df.index, y= "coefficients")
plt.xticks(rotation= 90);
```
